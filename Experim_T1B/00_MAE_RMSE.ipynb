```python
import pandas as pd

# MAE
def absolute_error(prevs, prevs_hat):
    #assert prevs.shape == prevs_hat.shape, 'wrong shape {prevs.shape} vs. {prevs_hat.shape}'
    return abs(prevs_hat - prevs).mean(axis=-1)

# RMSE
def relative_absolute_error(p, p_hat, eps=None):
    def __smooth(prevs, epsilon):
        n_classes = prevs.shape[-1]
        return (prevs + epsilon) / (epsilon * n_classes + 1)

    p = __smooth(p, eps)
    p_hat = __smooth(p_hat, eps)
    return (abs(p-p_hat)/p).mean(axis=-1)

# Preprocesado para calcular errores
def prepro(df):
    REALES=df['REAL'].to_numpy()
    PREDICHAS=df['PREDICHAS'].to_numpy()
    return REALES,PREDICHAS
```

### <font color="#CA7868"> Método LR </font>

```python
if AC.shape != OVR_AC.shape:
    print('AC y OVR_AC necesita preprocesado')
else:
    print('AC  y OVR_AC   tienen misma dimensión')
    
if HDY.shape != OVR_HDY.shape:
    print('HDY y OVR_HDY necesita preprocesado')  
else:
    print('HDY y OVR_HDY  tienen misma dimensión')    
    
if HDX.shape != OVR_HDX.shape:
    print('HDX y OVR_HDX necesita preprocesado')  
else:
    print('HDX y OVR_HDX  tienen misma dimensión')    
    
if EDX.shape != OVR_EDX.shape:
    print('EDX y OVR_EDX necesita preprocesado')  
else:
    print('EDX y OVR_EDX  tienen misma dimensión')        
```

## <font color="#CA7868"> MAE para LR </font>

```python
REALES_AC=prepro(AC)[0]
PREDICHAS_AC=prepro(AC)[1]
REALES_OVR_AC=prepro(OVR_AC)[0]
PREDICHAS_OVR_AC=prepro(OVR_AC)[1]

REALES_HDY=prepro(HDY)[0]
PREDICHAS_HDY=prepro(HDY)[1]
REALES_OVR_HDY=prepro(OVR_HDY)[0]
PREDICHAS_OVR_HDY=prepro(OVR_HDY)[1]

REALES_HDX=prepro(HDX)[0]
PREDICHAS_HDX=prepro(HDX)[1]
REALES_OVR_HDX=prepro(OVR_HDX)[0]
PREDICHAS_OVR_HDX=prepro(OVR_HDX)[1]


REALES_EDX=prepro(EDX)[0]
PREDICHAS_EDX=prepro(EDX)[1]
REALES_OVR_EDX=prepro(OVR_EDX)[0]
PREDICHAS_OVR_EDX=prepro(OVR_EDX)[1]


REALES_EDY=prepro(EDY)[0]
PREDICHAS_EDY=prepro(EDY)[1]
REALES_OVR_EDY=prepro(OVR_EDY)[0]
PREDICHAS_OVR_EDY=prepro(OVR_EDY)[1]


print('================= MAE =================')
print('AC     ',absolute_error(REALES_AC,PREDICHAS_AC))
print('OVR-AC ',absolute_error(REALES_OVR_AC,PREDICHAS_OVR_AC))
print(' ')
print('HDY    ',absolute_error(REALES_HDY,PREDICHAS_HDY))
print('OVR-HDY',absolute_error(REALES_OVR_HDY,PREDICHAS_OVR_HDY))
print(' ')
print('HDX    ',absolute_error(REALES_HDX,PREDICHAS_HDX))
print('OVR-HDX',absolute_error(REALES_OVR_HDX,PREDICHAS_OVR_HDX))
print(' ')
print('EDX    ',absolute_error(REALES_EDX,PREDICHAS_HDX))
print('OVR-EDX',absolute_error(REALES_OVR_EDX,PREDICHAS_OVR_EDX))
print(' ')
print('EDY    ',absolute_error(REALES_EDY,PREDICHAS_HDY))
print('OVR-EDY',absolute_error(REALES_OVR_EDY,PREDICHAS_OVR_EDY))
```

##### <font color="#CA7868"> RMSE para LR</font>

```python
print('================= RMSE =================')
print('AC     ',relative_absolute_error(REALES_AC,PREDICHAS_AC,eps=0.0005))
print('OVR-AC ',relative_absolute_error(REALES_OVR_AC,PREDICHAS_OVR_AC,eps=0.0005))
print(' ')
print('HDY    ',relative_absolute_error(REALES_HDY,PREDICHAS_HDY,eps=0.0005))
print('OVR-HDY',relative_absolute_error(REALES_OVR_HDY,PREDICHAS_OVR_HDY,eps=0.0005))
print(' ')
print('HDX    ',relative_absolute_error(REALES_HDX,PREDICHAS_HDX,eps=0.0005))
print('OVR-HDX',relative_absolute_error(REALES_OVR_HDX,PREDICHAS_OVR_HDX,eps=0.0005))
print(' ')
print('EDX    ',relative_absolute_error(REALES_EDX,PREDICHAS_EDX,eps=0.0005))
print('OVR-EDX',relative_absolute_error(REALES_OVR_EDX,PREDICHAS_OVR_EDX,eps=0.0005))
print(' ')
print('EDX    ',relative_absolute_error(REALES_EDY,PREDICHAS_EDY,eps=0.0005))
print('OVR-EDX',relative_absolute_error(REALES_OVR_EDY,PREDICHAS_OVR_EDY,eps=0.0005))
```
